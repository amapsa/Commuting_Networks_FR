```python
#introduce auxiliary function visualizing the degree rank plots
def rankPlot(x,fit=0,params=[]):
    #visualize the rankplot for the array x, fitting it with a power law, 
    #lognormal or normal distribution (according to fit) 
    #params - power-law fit cut-off
    
    xs=sorted(x,reverse=True) # sort the sequence in decreasing order
    plt.loglog(range(1, len(xs)+1),xs,marker='.') #create a log-log plot of the value vs its rank
    plt.title("Centrality rank plot") #plot title
    
    #axe labels  
    plt.ylabel("Centrality")
    plt.xlabel("rank")
    
    fp=[]  
    if fit==1: #power law fit
        xl=np.array(xs)  
        xl=log(xl[xl>0])
        r=log(np.array(range(1,len(xl)+1)))
        data=pd.DataFrame({'x':r,'y':xl})  
        if len(params)==0:
            params=[0, Inf]
        if len(params)==1:
            params=append(params,Inf)
        lm = smf.ols(formula='y ~ x'.encode('ascii'), data = data.loc[(x[x>0]>=params[0])&(x[x>0]<=params[1])]).fit()
        fp= lm.params.x
        plt.loglog(exp(data.x),exp(lm.predict(data)),'r',marker='', label='{0:.2f}'.format(fp))   
        
    if fit==2: #lognormal fit
        xl=log(x[x>0])
        sigma=std(xl)
        mu=mean(xl)  
        fp=sigma
        s=exp(mu)*stat.lognorm.ppf((np.array(range(0,len(xl)+1))+0.5)/(len(xl)+1),sigma)[::-1]
        plt.loglog(s,marker='')
        
    if fit==3: #normal fit
        xl=x[x>0]
        sigma=std(xl)
        mu=mean(xl)  
        fp=sigma
        s=stat.norm.ppf((np.array(range(0,len(xl)+1))+0.5)/(len(xl)+1),loc=mu,scale=sigma)[::-1]
        plt.loglog(s,'r',marker='')      
        
    return fp  
```

```python
#introduce a function visualizing degree partial and cumulative distributions
def distribution(x,cdf,params=10):
    xh, xb=np.histogram(log(x[x>1]),bins=params)
    xh=1.0*xh/len(x)
    xb=(xb[1:]+xb[0:-1])/2
    xb=xb[xh>0]
    xh=xh[xh>0]
    if cdf:
        xh=1-cumsum(xh[:-1])
        xb=xb[:-1]
        plt.ylabel("Log(P(c>=x))")
    else:
        plt.ylabel("Log(P(c=x))")
    plt.loglog(exp(xb),xh,marker='o')
    plt.xlabel("Log Degree Centrality, x")
```

## B. Gravity model w/ local normalization

```python
#Set the normalization parameter - here distance
nxB = DG
nx.set_edge_attributes(nxB,'Dist', 0)

for e in nxB.edges():
    nxB[e[0]][e[1]]['Dist'] = links[(links.O_UU_EPCI == e[0]) & (links.D_UU_EPCI  == e[1])]['Dist'].values[0]

nxB_wout=nxB.out_degree(weight='weight')
nxB_win=nxB.in_degree(weight='weight')
```

```python
#A few statistics on the degrees
a = pd.DataFrame.from_dict(dict(nxB_wout), orient='index').rename(columns={0: 'Out'})
b = pd.DataFrame.from_dict(dict(nxB_win), orient='index').rename(columns={0: 'In'})

print "\nAverage Out degree: {0:.4f}".format(a.mean()[0])
print "Min Out degree: {1}, Max Out degree: {0}".format(a.max()[0], a.min()[0])
print "Std Out Degree: {0:.4f}".format(a.std()[0])

print "\nAverage In degree: {0:.4f}".format(b.mean()[0])
print "Min In degree: {1}, Max In degree: {0}".format(b.max()[0], b.min()[0])
print "Std In Degree: {0:.4f}".format(b.std()[0])
```

$$
e(a,b)=T(a) \frac{w(b)^\beta d(a,b)^{-\gamma}}{\sum_{c\neq a} w(c)^\beta d(a,c)^{-\gamma}},
$$
Given the known total outgoing strength $T(a)$ for each node $a$,

```python
# function locally normalized gravity model fit (in terms of log-scale R2) for given exponents q
def LNGmodel(q):
    #first pass - assign relative weights as the model prediction
    for e in nxB.edges():
        if e[0] != e[1]:
            idx = links[(links.O_UU_EPCI == e[0]) & (links.D_UU_EPCI  == e[1])].index.values[0]
            nxB[e[0]][e[1]]['predict']=(nxB[e[0]][e[1]]['Dist']**q[0])*(links.D_ACT_TOTAL[idx]**q[1])
        else:
            nxB[e[0]][e[1]]['predict']=nxB[e[0]][e[1]]['weight']
            
    #compute normalization factors    
    nxB_norm=dict(nxB.out_degree(weight='predict'))
    for a in nxB.nodes():
        if nxB[a] != {}: 
            try:
                nxB_norm[a]=(dict(nxB_wout)[a]-nxB[a][a]['weight'])/(nxB_norm[a] - nxB[a][a]['predict'])
            except KeyError:
                nxB_norm[a]=(dict(nxB_wout)[a])/(nxB_norm[a])
                #except ZeroDivisionError: 
                 #   print a
    #second pass - normalize model predictions   
    E=np.array([0.0]*nxB.size())
    me=np.array([0.0]*nxB.size())
    i=0
    for e in nxB.edges():
        if e[0]!=e[1]:
            nxB[e[0]][e[1]]['predict']*=nxB_norm[e[0]]
            E[i]=nxB[e[0]][e[1]]['weight']
            me[i]=nxB[e[0]][e[1]]['predict']
            i+=1
    E=E[0:i]
    me=me[0:i]
    me=me[E>0]
    E=E[E>0]
    R2=1-sum(pow(log(E)-log(me),2))/sum(pow(log(E)-mean(log(E)),2))    
    return (R2,E,me) #performance and edge weights-predictions for fit visualization purposes 
    
def LNGmodelFit(q):
    #version for maximization
    return -LNGmodel(q)[0]
```

```python
#fit model parameters
res=optimize.minimize(LNGmodelFit,[0, 1],method='Nelder-Mead')
```

```python
(R2,e,me)=LNGmodel(res.x)
pl.figure(figsize=(8,6))
sns.regplot(log(me),log(e),fit_reg=False)
plt.plot(log(me),log(me),'r')  
pl.xlabel('Predicted Flow', fontsize=20)
pl.ylabel('Actual Flow', fontsize=20)
pl.savefig('Figures/local_normalization.jpg')
```

## A - Modules

```python
def modularity(G,partition):
    #compute network modularity according to the given partitioning
    nodes=G.nodes()
    #compute node weights and total network weight
    if G.is_directed():
        w1=G.out_degree(weight='weight')
        w2=G.in_degree(weight='weight')
        T=1.0*sum([e[2]['weight'] for e in G.edges(data=True)])
    else:
        w1=G.degree(weight='weight')
        w2=G.degree(weight='weight')
        T=1.0*sum([(1+(e[0]!=e[1]))*e[2]['weight'] for e in G.edges(data=True)])
    M=0 #start accumulating modularity score
    for a in nodes:
        for b in nodes:
            #if (G.is_directed())|(b>=a):
                if partition[a]==partition[b]: #if nodes belong to the same community
                    #get edge weight
                    if G.has_edge(a,b):
                        e=G[a][b]['weight']
                    else:
                        e=0
                    M+=e/T-w1[a]*w2[b]/(T**2) #add modularity score for the considered edge 
    return M                
```

```python
#remove loop edges from a network
def remove_loops(G):
    GU=G
    for n in GU.nodes():
        if GU.has_edge(n,n):
            GU.remove_edge(n,n)
    return GU  
```

```python
#produce symmetrized undirected version of a directed network
def as_undirected(G):
    GU=nx.Graph()
    for e in G.edges(data=True):
        if GU.has_edge(e[0],e[1]):
            GU[e[0]][e[1]]['weight']=GU[e[0]][e[1]]['weight']+e[2]['weight']
        else:
            GU.add_edge(e[0],e[1],weight=e[2]['weight'])
    return GU        
```

```python
#visualize network partitioning given node locations pos
def visualizePartition(G,partition,pos):
    N=len(G.nodes())
    s=4+4*int(log10(N))
    plt.figure(figsize=(s,s))
    PN=max(partition.values())
    my_cmap = matplotlib.cm.hsv(np.linspace(0,1,PN+1)) #create a colormap for a given number of communities
    c=[]
    for n in G.nodes():
        c.append(1.0*partition[n]/PN)
    nx.draw(G,pos=pos,with_labels=False,arrows=True,node_size=300,node_color=c,width=1,edge_color='black')
```

```python
#interface for running compiled combo over the network G given a maximal number of communities maxcom
def getComboPartition(G,maxcom):
    #save network in net format
    nodes={}
    nodenum={}
    i=0
    #create a dictionary transforming nodes to unique numbers
    for n in G.nodes():
        nodenum[n]=i
        nodes[i]=n
        i+=1
    f = open('combo/temp.net', 'w')
    f.write('*Arcs\n')
    for e in G.edges(data=True):
        f.write('{0} {1} {2}\n'.format(nodenum[e[0]],nodenum[e[1]],e[2]['weight']))
    f.close()
    #run combo
    command=workfolder+'/combo/comboCPP combo/mobil_fr.net'
    if maxcom<Inf:
        command=command+' {0}'.format(maxcom)  
    os.system(command)
    #read resulting partition
    f = open('combo/mobil_fr_comm_comboC++.txt', 'r')
    i=0
    partition={}
    for line in f:
        partition[nodes[i]]=int(line)
        i+=1
    return partition    
```

```python
#Attempt at visualization
#The forces between the largest areas are so strong that everything is pulled in the center of the graph
#Another visualization is necessary

pl.figure(figsize=(10,10))

#Dict with the large poles' positions set
fixed_positions = {'Marseille - Aix-en-Provence':(43.3,5.4),'Toulon':(43.1167,5.933), 'Avignon':(43.95,4.8167), 
                   'Nîmes':(43.83,4.35), 'Salon-de-Provence':(43.633,5.1), 'Arles':(43.6768, 4.6303)}

fixed_nodes = fixed_positions.keys()

pos = nx.spring_layout(Dsud,pos=fixed_positions, fixed = fixed_nodes)

#We could activate the labels only for the large poles, but they are all surimposed
nx.draw_networkx(Dsud,pos,with_labels=False,style='dotted',edge_color='green')
```
